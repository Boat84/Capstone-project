```python
import re
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
import os
```

```python
# preprocessing of text
def preprocess_text(text):
    text = text.replace('\n', ' ').replace('\r', ' ')
    text = text.strip()
    text = re.sub(r'\s([?.!"](?:\s|$))', r'\1', text)
    #text = re.sub(r'[^\w\s]', '', text)
    text = text.lower()
    return text


def classify_and_save_scenes(input_file, output_file):
    # open file 
    print(input_file)
    with open(input_file, 'r', encoding='utf-8') as f:
        text = f.read()

    #split the text into scenes
    scenes = text.split("==================================================")
    #print(len(scenes))
    #print(scenes[1])
    
    #preprocessing of text
    scenes_processed = []
    for scene in scenes:
        scene_p = preprocess_text(scene)
        scenes_processed.append(scene_p)

    #print(scenes_processed[1])
    # create analyser 
    analyzer = SentimentIntensityAnalyzer()

    # get the scores
    scene_scores = []
    i = 0
    for scene in scenes_processed:
        preprocessed_text = preprocess_text(scene)
        scores = analyzer.polarity_scores(preprocessed_text)
        scene_scores.append({
            #"Scene": scene,
            "Scene": i,
            "Negative": scores['neg'],
            "Neutral": scores['neu'],
            "Positive": scores['pos'],
            "Compound": scores['compound']
        })
        i += 1
    # create DataFrame
    df = pd.DataFrame(scene_scores)

    # save to CSV 
    df.to_csv(output_file, index=False)

    print("Sentiment scores saved to scene_sentiment_scores.csv")

```

```python

try:
    directory_path = '/Users/xiaozhouye/Desktop/neuefische/Capstone-project_copy2/data/screenplay_data/data/scene_separated_texts'#-------------change your path 
    files = os.listdir(directory_path)
    directory_path_new = '/Users/xiaozhouye/Desktop/neuefische/Capstone-project_copy2/data/screenplay_data/data/scene_separated_analysed_vader_new3'#---------change your path for the csv file with the sentiment_score
    #print(len(files))
    if not os.path.exists(directory_path_new):
        os.makedirs(directory_path_new)
    for file in files:
        file_old = os.path.join(directory_path,file)
        file_new = os.path.join(directory_path_new,file)
        file_new = file_new.replace('.txt','.csv')
        classify_and_save_scenes(file_old, file_new)
except Exception as e:
    print(f"Error occurred: {e}")
```
