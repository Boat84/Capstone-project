```python

import numpy as np
import pandas as pd
import os
import re
import csv
```

```python
#remove"dialog :" from BERT 
def remove_prefix(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        content = file.read()
    
    modified_content = content.replace('dialog:', '')
    
    with open(file_path, 'w', encoding='utf-8') as file:
        file.write(modified_content)

```

```python
def remove_all_prefix(dir):
    try:
        files = os.listdir(dir)
        for file in files:
            file_p = os.path.join(dir,file)
            remove_prefix(file_p)
    except Exception as e:
        print(f"An error occurred: {str(e)}")
            
```

```python

def get_name_id(file):
    import re
    #define the pattern to extract_name and id
    pattern = r"^(.*)_(\d+)_.*$"
    match = re.match(pattern, os.path.basename(file)) 
    if match:
        name = match.group(1)
        id_number = match.group(2)
        return name, id_number
    else:
        raise ValueError("String format is incorrect")
```

```python
#statistic sentences and words

def statistic_sentences_words(file):
    import nltk
    from nltk.tokenize import sent_tokenize, word_tokenize
    from nltk.corpus import stopwords
    import string
    nltk.download('punkt')
    nltk.download('stopwords')
    
    with open(file, 'r', encoding='utf-8') as file:
        content = file.read()

    # split sentences
    sentences = sent_tokenize(content)
    num_sentences = len(sentences)

    # split words and remove symbols
    words = word_tokenize(content)
    words = [word for word in words if word.isalpha()]

    num_words = len(words)

    return num_sentences, num_words



```

```python
def statistic_sentences_words_all(dir):
    name = []
    Id = []
    sentences = []
    words = []
    try:
        files = os.listdir(dir)
        for file in files:
            file_p = os.path.join(dir, file)
            a, b = get_name_id(file_p)
            c, d = statistic_sentences_words(file_p)
            name.append(a)
            Id.append(b)
            sentences.append(c)
            words.append(d)
        
        out_csv = os.path.join(dir, 'summ.csv')
        with open(out_csv, 'w', newline='') as csvfile:
            csv_writer = csv.writer(csvfile)
            csv_writer.writerow(['Name', 'Id', 'sentences', 'words'])
            for i in range(len(name)):
                csv_writer.writerow([name[i], Id[i], sentences[i], words[i]])
        
        print("Statistics saved to summ.csv")
    except Exception as e:
        print(f"An error occurred: {str(e)}")
```
