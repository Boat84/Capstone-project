```python
#remove stopwords and symbols:
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import numpy as np
import pandas as pd
import os
import re

# 下载必要的NLTK数据包
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

```

```python
def preprocess_text(text):
    # remove HTML tag
    text = re.sub(r'<.*?>', '', text)
    # remove all no apl
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # trans all to lowwercase
    text = text.lower()
    # tokenization
    words = word_tokenize(text)
    # remove the stops words
    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word not in stop_words]
    # lemmatization
    lemmatizer = WordNetLemmatizer()
    words = [lemmatizer.lemmatize(word) for word in words]
    
    return ' '.join(words)

```

```python
def perprocess_all(dir, new_dir):
    try:
        files = os.listdir(dir)
        for file in files:
            file_p = os.path.join(dir, file)
            with open(file_p, 'r', encoding='utf-8') as f:
                content = f.read()
                content_new = preprocess_text(content)
                
            file_p_new = os.path.join(new_dir, file) 
            
            with open(file_p_new, 'w', encoding='utf-8') as txtfile:
                txtfile.write(content_new)
                
            print(f"Preprocessed {file} and saved to {file_p_new}")
    
    except Exception as e:
        print(f"An error occurred: {str(e)}")
```

```python
import os
import re
import csv

def save_path(dir):
    names = []
    imdbids = []
    paths = []
    try:
        files = os.listdir(dir)
        for file in files:
            file_p = os.path.join(dir, file)
            pattern = r"^(.*)_(\d+)_.*$"
            match = re.match(pattern, os.path.basename(file))
            if match:
                name = match.group(1)
                id_number = match.group(2)
                names.append(name)
                imdbids.append(id_number)
                paths.append(file_p)
            else:
                raise ValueError("String format is incorrect")
        
        s_path_csv = os.path.join(dir, 's_path.csv')
        with open(s_path_csv, 'w', newline='') as csvfile:
            csv_writer = csv.writer(csvfile)
            csv_writer.writerow(['Name', 'Id', 'dir'])
            for i in range(len(names)):
                csv_writer.writerow([names[i], imdbids[i], paths[i]])
        
        print("Statistics saved to s_path.csv ")
    except Exception as e:
        print(f"An error occurred: {str(e)}")
```
